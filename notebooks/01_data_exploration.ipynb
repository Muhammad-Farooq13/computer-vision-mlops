# Computer Vision Project - Data Exploration

This notebook explores the dataset and provides insights into the data distribution, image characteristics, and potential preprocessing requirements.

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from pathlib import Path
import sys

# Add src to path
sys.path.append('..')

from src.data.data_loader import DataLoader
from src.visualization.visualize import plot_images_grid, plot_class_distribution

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (15, 10)

print("Libraries imported successfully!")
```

## 1. Load Dataset

```python
# Initialize data loader
data_loader = DataLoader('../data/raw')

# Get dataset statistics
stats = data_loader.get_dataset_stats()
print("Dataset Statistics:")
print(f"  Training samples: {stats.get('train', 0)}")
print(f"  Validation samples: {stats.get('valid', 0)}")
print(f"  Test samples: {stats.get('test', 0)}")
print(f"  Total samples: {sum(stats.values())}")
```

## 2. Load Class Information

```python
# Load annotations
train_annotations = data_loader.load_annotations('train')

if 'classes' in train_annotations:
    classes = train_annotations['classes']
    print(f"Number of classes: {len(classes)}")
    print(f"\nClasses: {classes}")
else:
    print("No class information found")
```

## 3. Visualize Sample Images

```python
# Load sample images from training set
sample_images = data_loader.load_images('train', limit=9)

if sample_images:
    plot_images_grid(sample_images, rows=3, cols=3, figsize=(15, 15))
    plt.suptitle('Sample Training Images', fontsize=16, fontweight='bold')
    plt.show()
else:
    print("No images found to visualize")
```

## 4. Analyze Image Properties

```python
# Analyze image dimensions and properties
if sample_images:
    heights = [img.shape[0] for img in sample_images]
    widths = [img.shape[1] for img in sample_images]
    channels = [img.shape[2] if len(img.shape) == 3 else 1 for img in sample_images]
    
    print("Image Properties:")
    print(f"  Height - Min: {min(heights)}, Max: {max(heights)}, Avg: {np.mean(heights):.2f}")
    print(f"  Width - Min: {min(widths)}, Max: {max(widths)}, Avg: {np.mean(widths):.2f}")
    print(f"  Channels: {set(channels)}")
    
    # Plot distribution
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    axes[0].hist(heights, bins=20, color='steelblue', edgecolor='black')
    axes[0].set_title('Image Height Distribution')
    axes[0].set_xlabel('Height (pixels)')
    axes[0].set_ylabel('Frequency')
    
    axes[1].hist(widths, bins=20, color='coral', edgecolor='black')
    axes[1].set_title('Image Width Distribution')
    axes[1].set_xlabel('Width (pixels)')
    axes[1].set_ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()
```

## 5. Analyze Color Distribution

```python
# Analyze color distribution
if sample_images:
    # Calculate mean color per channel
    mean_colors = []
    for img in sample_images:
        mean_b, mean_g, mean_r = cv2.mean(img)[:3]
        mean_colors.append([mean_r, mean_g, mean_b])
    
    mean_colors = np.array(mean_colors)
    
    # Plot
    fig, ax = plt.subplots(figsize=(10, 6))
    
    x = np.arange(len(sample_images))
    width = 0.25
    
    ax.bar(x - width, mean_colors[:, 0], width, label='Red', color='red', alpha=0.7)
    ax.bar(x, mean_colors[:, 1], width, label='Green', color='green', alpha=0.7)
    ax.bar(x + width, mean_colors[:, 2], width, label='Blue', color='blue', alpha=0.7)
    
    ax.set_xlabel('Image Index')
    ax.set_ylabel('Mean Pixel Value')
    ax.set_title('Mean Color Values per Image')
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

## 6. Recommendations for Preprocessing

Based on the analysis above, consider the following preprocessing steps:

1. **Image Resizing**: Standardize all images to a common size (e.g., 224x224)
2. **Normalization**: Normalize pixel values to [0, 1] range or use ImageNet statistics
3. **Data Augmentation**: Apply augmentation techniques to increase dataset diversity
4. **Color Space**: Consider converting to different color spaces if needed
5. **Handling Imbalance**: Check for class imbalance and apply appropriate techniques
